# Size Squirrel

Eine Kletterschuh-Datenbank mit regem Feedback aus der Community: https://sizesquirrel.com/

```python
df_squirrel = pd.read_csv('./data/sizesquirrel_ratings.csv')
```

```python
df_squirrel.info()
```

```python
df_squirrel.head()
```

```python
df_squirrel.drop(['web-scraper-order','web-scraper-start-url','pagination', 'pagination-href','shoe', 'shoe-href' ],axis=1,inplace=True)
df_squirrel.head()
```

```python
df_squirrel['gender'] = df_squirrel['gender'].map(lambda x: x.split("'")[0])
df_squirrel['brand'] = df_squirrel['brand'].map(lambda x: x.title())
df_squirrel['model'] = df_squirrel['model'].map(lambda x: x.upper())
df_squirrel['size_scale'] = df_squirrel['size_scale'].map(lambda x: x.split("\n")[0])
df_squirrel['shoe_model'] = df_squirrel['brand'] + ' - ' + df_squirrel['model']
df_squirrel.drop(['brand','model'],axis=1,inplace=True)

df_squirrel.head()
```

```python
df_squirrel = df_squirrel[['shoe_model','gender', 'rating', 'climb_shoe_size', 'size_scale', 'fit',
       'foot_shape', 'uid']]
df_squirrel.head()
```

```python
round(df_squirrel.isnull().sum()/len(df_squirrel)*100,2)
```

# Bergfreunde

Ein großer Online-Shop im deutschsprachigen Raum für Outdoor Equipment mit relativ großer Community: https://www.bergfreunde.de/


```python
df_bergfreunde = pd.read_csv('./data/bergfreunde-pagination_20210312.csv')
```

```python
df_bergfreunde.info()
```

```python
df_bergfreunde.drop(['web-scraper-order','web-scraper-start-url', 'pagination','pagination-href','shoe', 'shoe-href','rating-href'],axis=1,inplace=True)
df_bergfreunde.head()
```

```python
df_bergfreunde['brand'] = df_bergfreunde['brand'].map(lambda x: x.split(' -')[0])
df_bergfreunde['model'] = df_bergfreunde['model'].map(lambda x: x.split(' -')[0].upper())
df_bergfreunde.head(2)
```

```python
df_bergfreunde['shoe_model'] = df_bergfreunde['brand'] + ' - ' + df_bergfreunde['model']
df_bergfreunde.drop(['brand', 'model'],axis=1,inplace=True)
df_bergfreunde.head(2)
```

```python
df_bergfreunde = df_bergfreunde[['shoe_model', 'climb_style', 'material', 'closure', 'sole', 'asymmetry', 'downturn',
       'vorspann', 'terrain', 'foot_shape', 'best_fit', 'overall_rating',
       'popularity', 'rating', 'uid', 'date', 'review_title', 'review_text']]
df_bergfreunde.head()
```

```python
df_bergfreunde['shoe_model'].unique()
```

```python
round(df_bergfreunde.isnull().sum()/len(df_bergfreunde)*100,2)
```

# Umfrage

Eigens erhobene Daten durch eine Online-Umfrage: https://forms.gle/PTs57af6sLqk32Wx7

```python
df_survey_shoes = pd.read_csv('./data/shoe_survey.csv')
df_users = pd.read_csv('./data/body_features.csv')
```

```python
df_survey_shoes.info()
```

```python
df_survey_shoes.isnull().sum()
```

```python
df_survey_shoes = df_survey_shoes.dropna(subset=['shoe_brand'])
df_survey_shoes.reset_index(inplace=True, drop=True)
```

```python
for i in df_survey_shoes[df_survey_shoes['Modell nicht in der Liste?'].notna() == True].index:
    df_survey_shoes.at[i, 'shoe_model'] = df_survey_shoes['Modell nicht in der Liste?'].iloc[i]

df_survey_shoes[df_survey_shoes['Modell nicht in der Liste?'].notna() == True].head()
```

```python
df_survey_shoes.at[61,'shoe_model'] = 'Five Ten - DIAMOND'
df_survey_shoes.at[77,'shoe_model'] = 'Boreal - SATORI'
df_survey_shoes.at[86,'shoe_model'] = 'Evolv - SHAMAN 2'
df_survey_shoes.at[88,'shoe_model'] = 'Evolv - SHAMAN 2'
df_survey_shoes.at[125,'shoe_model'] = 'Red Chili - AMP'

df_survey_shoes.drop('Modell nicht in der Liste?',axis=1,inplace=True)
```

```python
df_umfrage = pd.merge(df_survey_shoes, df_users, how='left', on=['uid'], suffixes=('_left', '_right'))
df_umfrage.info()
```

```python
round(df_umfrage.isnull().sum()/len(df_umfrage)*100,2)
```
